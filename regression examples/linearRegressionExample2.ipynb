# Linear Regression in Python Example

```python
%matplotlib inline

import random
import numpy as np
import matplotlib.pyplot as plt
```

Generate some data and plot

```python
# make up some random data and plot:
N = 200
X = 2+ np.random.uniform(.01,20, size=N)
y = 2*X*np.random.uniform(2,18, size=N)

#Plot the data
plt.scatter(X, y, marker='o', c='b')
plt.title('Profits distribution')
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.grid()
```

```python
from sklearn import linear_model
reg = linear_model.LinearRegression()
reg.fit(X,y) 
```
